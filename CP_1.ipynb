For our __Exploratory Data Analysis (EDA)__, we'll take it in two main steps:

* __1. Univariate Analysis:__:
    - Here, we'll focus on one feature at a time to understand its distribution and range.
    
    
* __2. Bivariate Analysis:__:
    - In this step, we'll explore the relationship between each feature. This helps us figure out the importance and influence of each feature on the targeted outcome.

With these two steps, we aim to gain insights into the individual characteristics of the data and also how each feature relates to our main goal: __predicting the target variable__.

```python
# Get continuous columns from your DataFrame
continuous_features = df.select_dtypes(include=[np.number]).columns

# Set up the subplot grid
n_features = len(continuous_features)
n_rows = (n_features + 2) // 3  # Number of rows needed
fig, ax = plt.subplots(nrows=n_rows, ncols=3, figsize=(18, n_rows * 6))
ax = ax.flatten()

# Loop to plot histograms for each continuous feature
for i, col in enumerate(continuous_features):
    sns.histplot(df[col].dropna(), bins=15, kde=True, ax=ax[i], color='#ff826e', edgecolor='none', alpha=0.7)
    ax[i].set_title(f'{col}', fontsize=14)
    ax[i].set_xlabel('')
    ax[i].set_ylabel('')
    ax[i].grid(True)

# Turn off any unused subplots
for i in range(n_features, len(ax)):
    ax[i].axis('off')

plt.suptitle('Distribution of Continuous Variables', fontsize=24, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

#**Removing Outliers**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot boxplots for each numeric feature
numeric_columns = df.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(20, 15))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(len(numeric_columns) // 3 + 1, 3, i)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')
plt.show()
```

In summary, outliers in the METABRIC breast cancer dataset may contain significant clinical information that could enhance the robustness of predictive models.

Removing these outliers without careful consideration might result in the loss of crucial insights and potentially reduce model performance.

```python
# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply outlier removal
for column in numeric_columns:
    df = remove_outliers(df, column)
```


The provided code uses the Interquartile Range (IQR) method to identify and remove outliers from the dataset. While this approach effectively eliminates extreme values that could skew analysis, it may not be ideal for the METABRIC breast cancer dataset.

Outliers in medical datasets often represent rare and clinically significant cases that could offer valuable insights. Removing these outliers without careful consideration could lead to the loss of important information and potentially decrease the robustness of predictive models.

# Define X and Y variables

_1- Data Splitting:__ We'll first split our dataset into a training set and a test set. This ensures that we have a separate set of data to evaluate our model's performance, untouched during the training and preprocessing phases.

__2. Hyperparameter Tuning & Cross-Validation:__ Given our dataset's size, to make the most of the available data during the model training phase, we'll apply __Gridsearch CV on the training set for hyperparameter tuning__. This allows us to get a better sense of how our model might perform on unseen data, without actually using the test set. The test set remains untouched during this phase and is only used to evaluate the final model's performance.


By following this structured approach, we ensure a rigorous training process, minimize the risk of data leakage, and set ourselves up to get a realistic measure of our model's performance on unseen data.

```python
# Define the features (X) and the output labels (y)
X = df.drop('Overall Survival Status', axis=1)
y = df['Overall Survival Status']
```

```python
# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=31)
```

```python
plt.figure(figsize=(3, 3))
plt.bar(df['Overall Survival Status'].value_counts().index, df['Overall Survival Status'].value_counts().values)
plt.xlabel('Overall Survival Status')
plt.ylabel('Count')
plt.title('Distribution of Overall Survival Status Classes')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert categorical columns to numeric
categorical_columns = X_train.select_dtypes(include=['object']).columns.tolist()

label_encoders = {}
for column in categorical_columns:
    encode = LabelEncoder()
    X_train[column] = encode.fit_transform(X_train[column])
    X_test[column] = encode.transform(X_test[column])
    label_encoders[column] = encode
```

###***StandardScaler***

```python
# Standardize the features (scaling)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

###***DecisionTreeClassifier***

```python
d = DecisionTreeClassifier(random_state=31)
```

```python
pg = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'auto', 'sqrt', 'log2'],
}
grid_search = GridSearchCV(estimator=d, param_grid=pg, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
best = grid_search.best_estimator_
best.fit(X_train, y_train)
```

###Model Evaluation

```python
y_pred = best.predict(X_test)
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Plotting the Confusion Matrix
cm_lr = confusion_matrix(y_test, best.predict(X_test))

plt.figure(figsize=(5, 5))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap="Oranges", cbar=False, annot_kws={"size": 30})
plt.title('Decision Tree Confusion Matrix', fontsize=15)
plt.xlabel("Predicted Value", fontsize=14)
plt.ylabel("True Value", fontsize=14)
plt.show()
```

#***LogisticRegression***

```python
# Train a model
model = LogisticRegression(random_state=31)
```

```python
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs', 'saga'],
    'penalty': ['l2'],
    'max_iter': [100, 200, 300]
}
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best = grid_search.best_estimator_

print("Best Parameters:", best_params)
```

```python
y_pred = best.predict(X_test)
y_pred_proba = best.predict_proba(X_test)[:, 1]

# Classification report
print(classification_report(y_test, y_pred))

# ROC AUC score
auc_score = roc_auc_score(y_test, y_pred_proba)
print("AUC Score:", auc_score)
```
